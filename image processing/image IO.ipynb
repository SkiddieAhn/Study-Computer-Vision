## 1. image read, display

```python
import cv2

imageFile = './data/lena.jpg'
img = cv2.imread(imageFile) # 영상 입력 
print(img.shape) # 영상 픽셀 수, (흑백,컬러) 확인 

img2=cv2.imread(imageFile,0) # 영상을 그레이스케일로 입력 (0=cv2.IMREAD_GRAYSCALE)
print(img2.shape)

cv2.imshow('Lena color',img) #윈도우 표시 
cv2.imshow('Lena grayscale',img2)

cv2.waitKey() # 키보드 입력 대기
cv2.destroyAllWindows() # 윈도우 파괴 
```

## 2. image save

```python
imageFile = './data/lena.jpg'
img = cv2.imread(imageFile) # 영상 입력 

cv2.imwrite('./out/Lena.bmp',img) # 확장자를 바꿔서 영상 저장 
cv2.imwrite('./out/Lena.png',img)
cv2.imwrite('./out/Lena2.png',img , [cv2.IMWRITE_PNG_COMPRESSION,9])
cv2.imwrite('./out/Lena2.png',img , [cv2.IMWRITE_JPEG_QUALITY,90])
```

## 3. color image display

```python
from matplotlib import pyplot as plt

imageFile='./data/lena.jpg'
imgBGR=cv2.imread(imageFile)
plt.axis('off')

# opencv는 BGR로 영상 입력을 받음, plt는 RGB로 영상 입력을 받음
# 따라서 바꿔야 됨 
imgRGB=cv2.cvtColor(imgBGR,cv2.COLOR_BGR2RGB)
plt.imshow(imgRGB)
plt.show()
```

## 4. gray image display

```python
imgGray=cv2.imread(imageFile,cv2.IMREAD_GRAYSCALE)
plt.axis('off')

plt.imshow(imgGray, cmap="gray", interpolation='bicubic') # '보간' 기술 적용 
plt.show()
```

## 5. margin, image save

```python
imageFile = './data/lena.jpg'
imgGray = cv2.imread(imageFile, cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(4,4)) # figure(plot) size 지정 (4*4 pixel) / plot: 그림,그래프 등을 담는 공간   
plt.subplots_adjust(left=0, right=0.5, bottom=0, top=0.5) # margin(여백) 지정 
plt.imshow(imgGray, cmap='gray')

plt.axis('off')
plt.show()
```

## 6. subplot image display

```python
path='./data/'

imgBGR1=cv2.imread(path+'lena.jpg')
imgBGR2=cv2.imread(path+'apple.jpg')
imgBGR3=cv2.imread(path+'baboon.jpg')
imgBGR4=cv2.imread(path+'orange.jpg')

# 컬러 변환: BGR -> RGB
imgRGB1 = cv2.cvtColor(imgBGR1, cv2.COLOR_BGR2RGB)
imgRGB2 = cv2.cvtColor(imgBGR2, cv2.COLOR_BGR2RGB)
imgRGB3 = cv2.cvtColor(imgBGR3, cv2.COLOR_BGR2RGB)
imgRGB4 = cv2.cvtColor(imgBGR4, cv2.COLOR_BGR2RGB)

# 서브플롯 2행2열개 생성 ,서브 플롯을 담는 plot 사이즈는 5X5
fig, ax = plt.subplots(2,2,figsize=(5,5))

# 4개의 차트에서 영상을 봄 
ax[0][0].axis('off')
ax[0][0].imshow(imgRGB1,aspect='auto')

ax[0][1].axis('off')
ax[0][1].imshow(imgRGB2,aspect='auto')

ax[1][0].axis('off')
ax[1][0].imshow(imgRGB3,aspect='auto')

ax[1][1].axis('off')
ax[1][1].imshow(imgRGB4,aspect='auto')

plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0.05, hspace=0.05)
plt.show()
```
